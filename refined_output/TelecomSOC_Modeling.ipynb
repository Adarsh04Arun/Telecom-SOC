```python
# Ensure expected columns exist
required_cols = ["label_suspicious","timestamp","msg"]
for c in required_cols:
    if c not in df_1.columns:
        raise ValueError(f"Missing required column: {c}")

# Make label numeric ints
df_1["label_suspicious"] = pd.to_numeric(df_1["label_suspicious"], errors="coerce").fillna(0).astype(int)

# Parse timestamp (UTC ok)
df_1["timestamp"] = pd.to_datetime(df_1["timestamp"], errors="coerce", utc=True)

# Sort and split by time: last 20% = test
df_sorted = df_1.sort_values("timestamp").reset_index(drop=True)
cut_idx = int(math.floor(len(df_sorted) * 0.8))
train_df = df_sorted.iloc[:cut_idx].copy()
test_df  = df_sorted.iloc[cut_idx:].copy()

len(train_df), len(test_df), train_df["timestamp"].min(), test_df["timestamp"].max()
```

```python
text_col = "msg"
cat_cols  = [c for c in ["host","user","process","event_type","action","status","severity","protocol","service","tags"] if c in df_1.columns]
num_cols  = [c for c in ["rule_score","severity_num","hour","is_internal_src","is_internal_dst"] if c in df_1.columns]

X_train = train_df[[text_col] + cat_cols + num_cols].copy()
y_train = train_df["label_suspicious"].copy()

X_test  = test_df[[text_col] + cat_cols + num_cols].copy()
y_test  = test_df["label_suspicious"].copy()

print("Features:", [text_col] + cat_cols + num_cols)
```

```python
from sklearn.impute import SimpleImputer

# ColumnTransformer: text, categorical, numeric
pre = ColumnTransformer(
    transformers=[
        ("txt", TfidfVectorizer(max_features=30000, ngram_range=(1,2), dtype=np.float32), text_col),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse=True), cat_cols),
        ("num", "passthrough", num_cols),
    ],
    remainder="drop",
    verbose_feature_names_out=False
)

# Choose model
if HAVE_LGB:
    from lightgbm import LGBMClassifier
    base_model = LGBMClassifier(
        n_estimators=600,
        learning_rate=0.05,
        max_depth=-1,
        num_leaves=64,
        subsample=0.9,
        colsample_bytree=0.8,
        class_weight="balanced",
        n_jobs=-1,
        random_state=42
    )
else:
    base_model = HistGradientBoostingClassifier(
        max_depth=None,
        max_iter=400,
        learning_rate=0.08,
        max_leaf_nodes=63,
        class_weight="balanced",
        random_state=42
    )

pipe = Pipeline([
    ("pre", pre),
    ("clf", base_model)
])
pipe
```

```python
pipe.fit(X_train, y_train)

# Probabilities on test set
try:
    y_score = pipe.predict_proba(X_test)[:,1]
except Exception:
    # HistGradientBoosting has decision_function; map to [0,1] via sigmoid-ish
    from scipy.special import expit
    y_score = expit(pipe.decision_function(X_test))

y_pred = (y_score >= 0.5).astype(int)

metrics = {
    "roc_auc": float(roc_auc_score(y_test, y_score)),
    "pr_auc": float(average_precision_score(y_test, y_score)),
}
p, r, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="binary", zero_division=0)
cm = confusion_matrix(y_test, y_pred, labels=[0,1])
metrics.update({
    "precision": float(p),
    "recall": float(r),
    "f1": float(f1),
    "confusion_matrix": {"tn": int(cm[0,0]), "fp": int(cm[0,1]), "fn": int(cm[1,0]), "tp": int(cm[1,1])},
    "model": type(base_model).__name__,
    "n_train": int(len(X_train)),
    "n_test": int(len(X_test)),
})
metrics
```

```python
# Score full dataset in time order
X_all = df_sorted[[text_col] + cat_cols + num_cols].copy()

try:
    proba_all = pipe.predict_proba(X_all)[:,1]
except Exception:
    from scipy.special import expit
    proba_all = expit(pipe.decision_function(X_all))

preds = df_sorted[["timestamp","host","event_type","src_ip","dst_ip"]].copy()
preds["model_proba"] = proba_all

# Save
OUT_DIR = Path("model_out")
OUT_DIR.mkdir(parents=True, exist_ok=True)

preds_path = OUT_DIR / "predictions_local.csv"
preds.to_csv(preds_path, index=False, encoding="utf-8")

metrics_path = OUT_DIR / "training_metrics.json"
metrics_path.write_text(json.dumps(metrics, indent=2), encoding="utf-8")

preds_path, metrics_path
```
